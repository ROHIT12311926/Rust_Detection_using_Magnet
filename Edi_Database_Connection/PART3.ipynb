```python
import pandas as pd
import joblib

# Load your model
model = joblib.load("rust_detection_rf_pipeline.pkl")

# Load production excel data
df = pd.read_excel("production_data_testing.xlsx")

# Make sure column names match model training
df = df.rename(columns={"Magnitude": "Resultant"})

# Select only the required columns
input_df = df[['Material', 'Length', 'Width', 'X', 'Y', 'Z', 'Resultant']]

# Predict for all rows
preds = model.predict(input_df)

# Count occurrences
rusted_count = list(preds).count("RUSTED")
nonrusted_count = list(preds).count("NON-RUSTED")

# Final decision
if rusted_count > nonrusted_count:
    final_output = "RUSTED"
else:
    final_output = "NON-RUSTED"

print("\n===============================")
print("   FINAL MAJORITY PREDICTION   ")
print("===============================")
print("RUSTED count     :", rusted_count)
print("NON-RUSTED count :", nonrusted_count)
print("\nðŸ‘‰ FINAL OUTPUT:", final_output)
```
